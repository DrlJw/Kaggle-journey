```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 20, 50)
y1 = 2 * x
y2 = -1 * x +20

plt.figure(num=3, figsize=(8, 5))
plt.plot(x, y2, color = 'k',alpha=0.8)# alpha 透明度可以改
plt.plot(x, y1, color='k', linewidth=1,alpha=0.8)
plt.axhline(y=9.87, xmin=-50, xmax=50, linewidth=1, color = 'k',alpha=0.8)
plt.axvline(x=3.2, ymin=-50, ymax=50, linewidth=1, color = 'k',alpha=0.8)
# plt.legend(loc='best')
plt.xlim((-4, 13))# xy的范围
plt.ylim((6, 18))
plt.xlabel('I am x')
plt.ylabel('I am y')
plt.title('title')
ax = plt.gca()
ax.spines['left'].set_position(('data',0))
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.grid(True, linestyle = "-.", color = "k", linewidth = 1, alpha=0.1)# 网格线
plt.scatter([10, ], [10, ], s=50, color='b')
plt.fill_between(x,y1,y2,color='blue',alpha=0.25)
plt.show()
```

```python
# 柱状图
n = 12
X = np.arange(n)
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)

plt.bar(X, +Y1, edgecolor='black')
plt.bar(X, -Y2, edgecolor='black')

plt.xlim(-.5, n)
plt.ylim(-1.25, 1.25)
plt.show()
```

```python
#correlation heatmap of dataset
def correlation_heatmap(df):
    _ , ax = plt.subplots(figsize =(14, 12))
    colormap = sns.diverging_palette(220, 10, as_cmap = True)
    
    _ = sns.heatmap(
        df.corr(), 
        cmap = colormap,
        square=True, 
        cbar_kws={'shrink':.9 }, 
        ax=ax,
        annot=True, 
        linewidths=0.1,vmax=1.0, linecolor='white',
        annot_kws={'fontsize':12 }
    )
    
    plt.title('Pearson Correlation of Features', y=1.05, size=15)

correlation_heatmap(pd.concat([train.iloc[:,2:6],train.iloc[:,1]],axis=1))
```
